## Datum in Alter umwandeln

```python
df['birthday'] = pd.to_datetime(df['birthday'], format='%m/%d/%Y')
```

```python
df.info()
```

```python
df.sort_values('birthday',ascending=True)
```

```python
now = pd.Timestamp('now')
df['age'] = (now - df['birthday']).astype('<m8[Y]')
```
